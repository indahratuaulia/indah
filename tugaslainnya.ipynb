```
#nilai pathloss
def cos(frek,jenis,tt,hr,d):
  #frek = frekuensi (MHz)
  #jenis = urban, suburban, open
  #tt = tinggi tower (meter)
  #hr = tinggi user (meter)
  #d = jarak (meter)
    if frek in range(150,1501):
        Ca=69.55
        Cb=26.16
    elif frek in range(1501,2001):
        Ca=46.3
        Cb=33.9
    
    if jenis=='urban' :
        Cm=0
    elif jenis=='suburban' :
        Cm= -2*((np.log10((frek/28)**2))-5.4)
    elif jenis=='open' :
        Cm= -4.78*(np.log10(frek))+18.33*(np.log10(frek))-40.94
    else :
        Cm='jenis salah'
    
    ahr = (1.1*np.log10(frek)-0.7)*hr-(1.56*np.log10(frek)-1.8)
    Lp = Ca+(Cb*np.log10(frek))-(13.83*np.log(tt))-ahr+((44.9-6.55*np.log10(tt))*np.log10(d))-Cm
    
    return Lp

#nilai daya terima
def daya_terima(ptx,Lp, gtx, grx):
  prx = 10*np.log10(ptx) - Lp + gtx + grx
  
  return 10**(prx/10)

#nili signal to noise ratio
def s2nr (bw, prx):
  no = -174 + 10*np.log10(bw)
  no1 = 10**(no/10)
  
  return prx/no1

#kapasitas seluler
def kapasitas_seluler (bw, snr):
  ks= bw * np.log2(1+snr)

  return ks
```

DATA

```

tt = 80 #meter
hr = 1.5 #meter
jenis = ['urban', 'suburban', 'open']
f = [ 700, 900, 1800]
jarakmax = 1000 #meter
jarakmin = 300 #meter
titik = 200 #user
d = np.random.randint(jarakmin, jarakmax, titik)
d = d[np.argsort(d)]
ptx = 30 #watt
gtx = 10 #dB
grx = 2 #dB
bw = 5000000 #Hz
```

Penghitungan fuensi 1800MHz di jenis 3 area

```
lp1 = []
prx1 = []
snr1 = []
ks1 = []
for i in range(3):
  lp1.append(cos(f[2], jenis[i], tt, hr, d))

  prx1.append(daya_terima(ptx, lp1[i], gtx, grx))
  prx_sort = prx1[i]
  prx1[i] = prx_sort[np.argsort(d)]

  snr1.append(s2nr(bw, prx1[i]))
  snr_sort = snr1[i]
  snr1[i]= snr_sort[np.argsort(d)]

  ks1.append(kapasitas_seluler(bw, snr1[i]))
  ks_sort = ks1[i]
  ks1[i]= ks_sort[np.argsort(d)]
```

Penghitungan data di tiap urban area dengan fuensi 700MHz, 900MHz, 1800MHz

```
lp2 = []
prx2 = []
snr2 = []
ks2 = []
for i in range(3):
  lp2.append(cos(f[i], jenis[0], tt, hr, d))

  prx2.append(daya_terima(ptx, lp2[i], gtx, grx))
  prx_sort1 = prx2[i]
  prx2[i] = prx_sort1[np.argsort(d)]

  snr2.append(s2nr(bw, prx2[i]))
  snr_sort1 = snr2[i]
  snr2[i]= snr_sort1[np.argsort(d)]

  ks2.append(kapasitas_seluler(bw, snr2[i]))
  ks_sort1 = ks2[i]
  ks2[i]= ks_sort1[np.argsort(d)]
```
